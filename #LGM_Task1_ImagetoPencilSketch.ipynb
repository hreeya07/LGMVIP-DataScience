## Image to Pencil Sketch with Python

Step1: Import these libraries

```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

Step2: Reading image and save its in an img vairiable

```python
img = cv2.imread('house img.jpg')
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

Step3: Convert the image from BGR to RGB

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Step4: Convert image from BGR to Gray Scale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Image", img_gray)
cv2.waitKey(0)
```

Step5: Invert the image

```python
img_invert = cv2.bitwise_not(img_gray)
cv2.imshow("Inverted Image", img_invert)
cv2.waitKey(0)
```

Step6: Blur the image by GaussianBlur Function

```python
img_blur = cv2.GaussianBlur(img_invert, (115,115), 0)
```

Step7: Using bitwise_not function to invert the blurred image

```python
img_blur_invert = cv2.bitwise_not(img_blur)
```

Step8: Convrting the blur image into Sketch image

```python
img_sketch = cv2.divide(img_gray, img_blur_invert, scale = 256.0)
```

Step9: Plotting the Original image and the Pencil Sketch Outputs

```python
plt.figure(figsize = (16,10))
plt.subplot(1,2,1)
plt.title('Original Image', size = 20)
plt.imshow(img_RGB)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Pencil Sketch', size = 20)
sketch_rgb = cv2.cvtColor(img_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(sketch_rgb)
plt.axis('off')

plt.show()
```
